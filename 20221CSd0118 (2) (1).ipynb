```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x));
ax.plot(x, np.cos(x));
ax.plot(x, np.tan(x));
ax.set_title('Simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label');
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x), label = 'sin', linestyle = 'dashed')
ax.plot(x, np.cos(x), label = 'cos', linestyle = 'dotted')
ax.plot(x, np.sin(x), label = 'sin x+1', linestyle = 'dashed')
ax.set_title('Multiple lines')
ax.set_xlabel('perid')
ax.set_ylabel('amplitude');
ax.legend()
plt.show();
```

```python
data = {'Apple': 10, 'Orange': 15, 'Lemon': 5, 'Lime':20}
names = list(data.keys())
values = list(data.values())
values.sort()
fig = plt.figure()
ax = plt.axes()
ax.barh(names, values);
```

```python
data = {'Apple': 10,'orange':15,'Bannana': 13, 'Pear': 9,'Pomogranate': 11,'Lemon': 5,'Lime': 20}
names = list(data.keys())
values = list(data.values())
values.sort()
ax = plt.axes()
ax.barh(names, values)
ax.set_xlabel('Quantity')
ax.set_ylabel('Fruit')
ax.set_title('Fruit Quantities');
```

```python
labels = ['G1', 'G2', 'G2', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means,width, label='Men')
rects2 = ax.bar(x + width/2, women_means,width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

ax.set_ylim(0,40);
```
